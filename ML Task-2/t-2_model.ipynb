# *TASK-2*: **Creating a K-MEANS clustering algorithm for Customer Segementation**

Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

Loading the Mall Customer Segmentation Dataset and Selecting "Annual Income (k$)" and "Spending Score (1-100)" as the features for tthis segmentation

```python
df=pd.read_csv("Mall_Customers.csv")

X=df[["Annual Income (k$)", "Spending Score (1-100)"]].values
```

## *Choosing the number of clusters using **Elbow Joint Method***

Finding the inertia values for each number of clusters from 1-10

```python
kr=range(1,11)
dist=[]
for k in kr:
  km=KMeans(n_clusters=k, random_state=42)
  km.fit(X)
  dist.append(km.inertia_)
```

Plotting the inertia values with the range and finding the "Elbow Joint"

```python
plt.figure(figsize=(8,6))
plt.plot(kr,dist, 'o:k')
plt.grid()
plt.xlabel('No. of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal K')
plt.show()
```

Here we can see that at No. of clusters=5 we can see the 'elbow joint'. So the no. of clusters that we take for clustering this dataset is **K=5** 

## *Implementing **K-Means** Clustering on the dataset*

Now with 5 clusters, lets fit and predict the kmeans on the dataset and get the centroids of all the 5 clusters

```python
kmeans=KMeans(n_clusters=5, random_state=42)

res=kmeans.fit_predict(X)
cluscen=kmeans.cluster_centers_
```

Plotting the Clusters and their centroids with Annual Income and Spending Score

```python
plt.figure(figsize=(8,6))
plt.scatter(X[:,0],X[:,1],c= res,cmap='viridis')
plt.scatter(cluscen[:,0],cluscen[:,1],s=200,c='red',marker='+')
plt.title('Customer Segmentation using K-Means Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

In the above Scatter plot  the 5 CLusters represents 5 segments of Customers with:

1. Low Income and Low Spending Score
2. Low Income and High Spending Score
3. Medium Income and Medium Spending Score
4. High Income and Low Spending Score
5. High Income and High Spending Score
